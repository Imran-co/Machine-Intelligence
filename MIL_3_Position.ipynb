```python
import pandas as pd


data=pd.read_csv("/content/Position_Salaries.csv")
print(data.head())
print(data.describe())
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['Position'],data['Salary'],)
plt.title('relationship between position and salary')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(data['Level'],data['Salary'],)
plt.title('Relationship between Level and Salary')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
x=data[['Level']]
y=data['Salary']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
y_predict=model.predict(x)
```

```python
plt.scatter(x,y, color='blue', label='Data points')
plt.plot(x,y_predict, color='red', label='regression line')
plt.title('linear regression')
plt.xlabel('levels of the employee')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
Poly=PolynomialFeatures(degree=3)
X_p=Poly.fit_transform(x)
model_p=LinearRegression()
model_p.fit(X_p,y)

y_predict_p=model_p.predict(X_p)
X_p=Poly.fit_transform(x)
model_p=LinearRegression()
model_p.fit(X_p,y)

y_predict_p=model_p.predict(X_p)
```

```python
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, y_predict_p, color='green', label='Polynomial regression line')
plt.title('Salary vs Levels (Polynomial Regression with degree 3)')
plt.xlabel('Levels of the employee')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
from sklearn.metrics  import mean_squared_error
mse = mean_squared_error(y, y_predict)
print("Mean Sqyare Error : ",mse)
```

```python
from sklearn.metrics import r2_score
R2 = r2_score(y, y_predict)
print("R-squared : ", R2)
```
