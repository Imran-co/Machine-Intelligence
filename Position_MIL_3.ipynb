```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error


data=pd.read_csv("/content/Position_Salaries.csv")
print(data.head())
print(data.describe())


plt.scatter(data['Position'],data['Salary'],)
plt.title('relationship between position and salary')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()


plt.scatter(data['Level'],data['Salary'],)
plt.title('relationship between level and salary')
plt.xlabel('level')
plt.ylabel('Salary')
plt.show()


x=data[['Level']]
y=data['Salary']

model=LinearRegression()
model.fit(x,y)
y_predict=model.predict(x)

plt.scatter(x,y, color='blue', label='Data points')
plt.plot(x,y_predict, color='red', label='regression line')
plt.title('linear regression')
plt.xlabel('levels of the employee')
plt.ylabel('Salary')
plt.legend()
plt.show()


p=PolynomialFeatures(degree=3)
x_p=p.fit_transform(x)
model_p=LinearRegression()
model_p.fit(x_p,y)


y_predict_p=model_p.predict(x_p)


plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, y_predict_p, color='green', label='Polynomial regression line')
plt.title('Salary vs Levels (Polynomial Regression with degree 3)')
plt.xlabel('Levels of the employee')
plt.ylabel('Salary')
plt.legend()
plt.show()


r2=r2_score(y,y_predict_p)
mse=mean_squared_error(y,y_predict_p)

print("r2=",r2)
print("mse=",mse)


p=PolynomialFeatures(degree=2)
x_p=p.fit_transform(x)
model_p=LinearRegression()
model_p.fit(x_p,y)


y_predict_p=model_p.predict(x_p)

plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, y_predict_p, color='green', label='Polynomial regression line')
plt.title('Salary vs Levels (Polynomial Regression with degree 2 )')
plt.xlabel('Levels of the employee')
plt.ylabel('Salary')
plt.legend()
plt.show()


r2=r2_score(y,y_predict_p)
mse=mean_squared_error(y,y_predict_p)

print("r2=",r2)
print("mse=",mse)
```
