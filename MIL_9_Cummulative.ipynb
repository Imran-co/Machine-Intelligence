```python
import numpy as np
import pandas as pd
mu = 0
sigma = 1
def pdf(x,mu,sigma):
  return(1/(sigma * np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mu)/sigma)**2)


low_bound =-10
dx=0.001

def cdf(x_values, mu, sigma=1, low_bound=-10, dx=0.001):
  cdf_value = []
  for x in x_values:
    points = np.arange(low_bound, x, dx)
    normal_pdf = pdf(points, mu, sigma)
    integral = np.trapz(normal_pdf, points)
    cdf_value.append(integral)
  return np.array(cdf_value)

x_values = np.linspace(-4, 4, 200)

cdf_value = cdf(x_values, mu, sigma, low_bound,)


x_values = np.linspace(-4, 4, 200)

cdf_value = cdf(x_values, mu, sigma, low_bound)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(x_values, cdf_value, label=f'CDF of N({mu}, {sigma}^2)')
plt.title('Cumulative Distribution Function (CDF) of Normal Distribution')
plt.xlabel('x')
plt.ylabel('CDF(x)')
plt.grid(True)
plt.legend()
plt.show()
```
